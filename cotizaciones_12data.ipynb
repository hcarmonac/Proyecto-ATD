```python
import requests

api_key = "ea51535a06ab42f0824812f815f2eb08" 

url = "https://api.twelvedata.com/quote?symbol=AAPL&apikey=" + api_key

response = requests.get(url)

print(f"Status Code: {response.status_code}")
```

```python
data = response.json()

precio_actual = float(data["close"])
nombre_empresa = data["name"]
cambio_porcentual = float(data["percent_change"])

maximo_anual = float(data["fifty_two_week"]["high"])
minimo_anual = float(data["fifty_two_week"]["low"])

print(f"{nombre_empresa} cotiza a ${precio_actual:.2f}")

if cambio_porcentual >= 1:
    print(f'{nombre_empresa} se encuentra en tendencia alcista a ${maximo_anual - precio_actual:2.f} del máximo anual.')
else:
    print(f'{nombre_empresa} se encuentra en tendencia bajista a ${precio_actual - minimo_anual:.2f} del mínimo anual.')
```
