```python
import requests
import plotly.graph_objects as go

def graficar_cotizacion(ticker, aos=1):
    # Adquisicin
    api_key = 'ea51535a06ab42f0824812f815f2eb08' 
    ticker = ticker
    outputsize = 252 * aos

    url = 'https://api.twelvedata.com/time_series'

    params = {
        'symbol': ticker,
        'interval': '1day',
        'outputsize': outputsize, 
        'apikey': api_key
    }

    response = requests.get(url, params=params)

    # Limpieza
    data = response.json()
    historico = data['values']
        
    fechas = [punto['datetime'] for punto in historico][::-1]
    precios = [float(punto['close']) for punto in historico][::-1]

    # Representacin
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=fechas, 
        y=precios,
        mode='lines',
        name=ticker,
        line=dict(color='royalblue', width=2),
        hovertemplate='<b>Precio:</b> $%{y:.2f}<extra></extra>'
    ))

    if aos==1:
        title = f'Evolucin de {ticker} en {aos} ao'
    else:
        title = f'Evolucin de {ticker} en {aos} aos'

    fig.update_layout(
        title=title,
        xaxis_title='Fecha',
        yaxis_title='Precio de Cierre ($)',
        hovermode='x unified', 
        template='plotly_white',
        hoverlabel=dict(
            bgcolor='white',
            font_size=13,
            font_family='Arial'
        )
    )

    fig.show()

graficar_cotizacion('NVDA', 5)
```
