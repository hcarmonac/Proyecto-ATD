```python
from socket import *
#Cliente
destino = input('Introduce el destino: ')
serverName = destino
serverPort = 10000
clientSocket = socket(AF_INET, SOCK_STREAM)
print('Cliente conectado')
clientSocket.connect((serverName,serverPort))

while True:
    print("-" * 50)
    print("  0 para introducir el nombre de la empresa")
    print("  1 EXIT")
    print("-" * 50)
    
    opcion = input("Selecciona una opción: ").strip()
    if opcion == '0':
        empresa = input('Introduce el nombre de la empresa: ')
        clientSocket.send(empresa.encode())
    elif opcion =='1':
        print('Programa finalizado')
        clientSocket.close()
        break
```
