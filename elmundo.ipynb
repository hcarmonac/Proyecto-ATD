```python
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.support.ui import Select
#PROBLEMA, si el nombre de la empresa que da el usuario es Nestle, si en el título o subtítulo es nestle con acento no lo coge
import time
#Introducir empresa
def elmundo (empresa):
    
    #Búsqueda elmundo.es
    try:
        driver = webdriver.Firefox()
        driver.get('https://www.elmundo.es/')
        driver.maximize_window()
    
    except:
        print('Url mal escogida')
    
    #Aceptar cookies
    try:
        cookies_button=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.ID, "ue-accept-notice-button"))
                )
        cookies_button.click()
    except:
        print('Error con el botón de cookies')
    
    #Clicar botón de búsqueda
    try:
        search_button=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.CLASS_NAME, "ue-c-main-header__search-box"))
                )
        search_button.click()
    except:
        print('Error con el botón de búsqueda')
    
    #Clicar búsqueda avanzada
    try:
        busqueda_avanzada=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.LINK_TEXT, "búsqueda avanzada »"))
                )
        busqueda_avanzada.click()
    except:
        print('Error con la búsqueda avanzada')
        
    #Elegir 50 resultados por página
    try:
        WebDriverWait(driver, 5).until(
                    EC.presence_of_all_elements_located((By.CLASS_NAME, 'consejos')) #Esperamos que carguen todas las opciones
                )
        desplegables = driver.find_elements(By.CLASS_NAME, 'consejos') #Cogemos todos los select, su clase es consejos
        select = Select(desplegables[2]) # Cogemos el tercer
        select.select_by_value("50")
    except: 
        print('Error seleccionando el número de resultados por página')

    #Elegir noticias del último año
    try:
        WebDriverWait(driver, 5).until(
                    EC.presence_of_all_elements_located((By.CLASS_NAME, 'consejos')) #Esperamos que carguen todas las opciones
                )
        select = Select(desplegables[0]) # Cogemos el tercer
        select.select_by_value("365")
    except: 
        print('Error seleccionando el número de resultados por página')
        
    # Especificamos el nivel de coincidencia de la noticia en mínimo 80%
    try:
        select = Select(desplegables[3]) # Cogemos el tercer
        select.select_by_value("70")
    except: 
        print('Error seleccionando el porcentaje de coincidencia')
        
    # Send keys del nombre de la empresa y enviar
    try:
        insertar_nombre=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.ID, "q"))
                )
        insertar_nombre.send_keys(empresa)
        insertar_nombre.send_keys(Keys.ENTER) #El enter envía el formulario
    
    except:
        print('Error con send keys del nombre de la empresa')
        
    #Clicamos un botón para ordenar las noticias por fecha
    try:
        boton_ordenar_fecha=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.LINK_TEXT, "Ordenar por FECHA"))
                )
        boton_ordenar_fecha.click()
    except:
        print('Error con el botón de ordenar por fecha')

    #No creo que sea necesario restringir a noticias de economía si solo usamos el último año
   # Clicar en economía para que solo muestre noticias de economía
    try:
        boton_economia=WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.LINK_TEXT, "Economía"))
                )
        boton_economia.click()
    except:
        print('Error con el botón de noticias de economía')
    
    lista=[]
    while True: # Que recoja los datos de cada página y luego clique en el botón de siguiente página, si existe; si no existe que cierre el bucle con break
        try:
            WebDriverWait(driver, 5).until(
                EC.presence_of_all_elements_located((By.TAG_NAME, "li")) #Esperamos a que carguen todos los li
            )
            
            todos_los_li = driver.find_elements(By.TAG_NAME, "li")
        
            # Guardamos solo las noticias que tengan el nombre de la empresa en su texto (título o subtítulo)
        except:
            print('Error cogiendo los li de las noticias')
        
    # Filtrar noticias importantes y guardar fechas (coherencia 80% o más)
        try:
            for i in todos_los_li:
                if empresa.lower() in i.text.lower():
                    elemento_a = i.find_element(By.TAG_NAME, 'a')
                    titulo = elemento_a.text
                    enlace = elemento_a.get_attribute('href')
                    fecha = i.find_element(By.CLASS_NAME, 'fecha').text
                    lista.append((fecha, titulo, enlace))
        except:
            print('Error filtrando que el nombre de la empresa este en título o subtítulo y guardándolo')
            
    #Ir a la siguiente página    
        try:
            boton_siguiente_pag=WebDriverWait(driver, 2).until(
                        EC.element_to_be_clickable((By.LINK_TEXT, "Siguiente »"))
            )
            
            boton_siguiente_pag.click()
        except:
            break
    return lista
```

```python
if __name__ == "__main__":
    nombre_empresa = str(input('Introduce el nombre de la empresa a buscar: '))
    lista = elmundo(nombre_empresa)
```
